```python
from textblob import TextBlob
```

```python
xx=df[0:73995]
```

```python
polarity=[]
for data in df['im getting on borderlands and i will murder you all ,']:
    try:
        polarity.append(TextBlob(data).sentiment.polarity)
    except:
        polarity.append(0)
```

```python
polarity[0:10]
```

```python
df['polarity']=polarity
```

```python
xx=df.head(20)
```

```python
xx.head()
```

```python
pos_comt=df[df['polarity']==0.5]
```

```python
pos_comt.head()
```

```python
neg_cmt=df[df['polarity']==-1]
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
cloud = WordCloud(stopwords=set(STOPWORDS)).generate(total)
plt.figure(figsize=(15, 10))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
